# Quick Data inspection

We have 4 drillholes DH7, DH8, DH13 and DH14.
From old to new they are DH7 -> DH8 -> DH13 -> DH14.

The units in the Model, (from young to old) are:  
* Quaternary  
* Andesite  
* Rhyolite  (with dykes of Qtz_Porphyryie)  
* Qz-Rhyolite 
* Dacite (partly heavily altered)  
* Rhyodacite  

Note, that the last entry in the boreholes is the End of the drillhole, NOT necessarily the layer-transition!

```python
import gempy as gp
import numpy as np
```

```python
geo_model = gp.create_model('POC_model')

gp.init_data(geo_model, [0, 500., 0., 200., -200, 200.], [50, 50, 50],
            path_i = '../data/all_boreholes.csv'
            );
```

```python
np.random.seed(100)
geo_model.set_topography(source='random', d_z=[120, 180], fd=1.8)
```

```python
gp.plot_2d(geo_model, show_data=True, show_topography=True)
```

```python
gp.plot_3d(geo_model, show_topography=True)
```
